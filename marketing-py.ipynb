```python
import pandas as pd
```

```python
df = pd.read_csv('/kaggle/input/analyze-the-marketing-spending/Marketing.csv')
```

```python
# TASK 1 - 

#Question 1.1: What is CTR for google_wide campaign on February 1, 2021?
#Question 1.2: What is Conversion 1 for instagram_tier2 campaign on February 5, 2021?
#Question 1.3: What is Conversion 2 for facebook_tier1 campaign on February 9, 2021?
#Question 1.4: What is AOV for facebook_lal campaign on February 13, 2021?
#Question 1.5: What is CPC for facebook_retargeting campaign on February 23, 2021?
#Question 1.6: What is CAC for the youtube_blogger campaign on February 24, 2021?
#Question 1.7: What is ROMI for the banner_partner campaign on February 28, 2021?

#Calculating Metrics
df['CTR'] = (df['clicks'] / df['impressions']) * 100
df['Conv_1'] = df['leads'] / df['clicks']
df['Conv_2'] = df['orders'] / df['leads']
df['AOV'] = df['revenue'] / df['orders']
df['CPC'] = df['mark_spent'] / df['clicks']
df['CAC'] = df['mark_spent'] / df['orders']
df['ROMI'] = (df['revenue'] - df['mark_spent']) / df['mark_spent']

# Selecting only the desired columns
output_df = df[['CTR', 'Conv_1', 'Conv_2', 'AOV', 'CPC', 'CAC', 'ROMI']]

# Displaying the DataFrame with selected columns

# Question 1.1: Displaying the CTR for the google_wide campaign on February 1, 2021
print("CTR for google_wide campaign on February 1, 2021 is:", output_df.loc[output_df.index[(df['campaign_name'] == 'google_wide') & (df['c_date'] == '2021-02-01')], 'CTR'].iloc[0])

# Question 1.2: Conversion 1 for instagram_tier2 campaign on February 5, 2021
conversion_1_instagram_tier2 = output_df.loc[output_df.index[(df['campaign_name'] == 'instagram_tier2') & (df['c_date'] == '2021-02-05')], 'Conv_1'].iloc[0]
print("Conversion 1 for instagram_tier2 campaign on February 5, 2021 is:", conversion_1_instagram_tier2)

# Question 1.3: Conversion 2 for facebook_tier1 campaign on February 9, 2021
conversion_2_facebook_tier1 = output_df.loc[output_df.index[(df['campaign_name'] == 'facebook_tier1') & (df['c_date'] == '2021-02-09')], 'Conv_2'].iloc[0]
print("Conversion 2 for facebook_tier1 campaign on February 9, 2021 is:", conversion_2_facebook_tier1)

# Question 1.4: AOV for facebook_lal campaign on February 13, 2021
aov_facebook_lal = output_df.loc[output_df.index[(df['campaign_name'] == 'facebook_lal') & (df['c_date'] == '2021-02-13')], 'AOV'].iloc[0]
print("AOV for facebook_lal campaign on February 13, 2021 is:", aov_facebook_lal)

# Question 1.5: CPC for facebook_retargeting campaign on February 23, 2021
cpc_facebook_retargeting = output_df.loc[output_df.index[(df['campaign_name'] == 'facebook_retargeting') & (df['c_date'] == '2021-02-23')], 'CPC'].iloc[0]
print("CPC for facebook_retargeting campaign on February 23, 2021 is:", cpc_facebook_retargeting)

# Question 1.6: CAC for the youtube_blogger campaign on February 24, 2021
cac_youtube_blogger = output_df.loc[output_df.index[(df['campaign_name'] == 'youtube_blogger') & (df['c_date'] == '2021-02-24')], 'CAC'].iloc[0]
print("CAC for the youtube_blogger campaign on February 24, 2021 is:", cac_youtube_blogger)

# Question 1.7: ROMI for the banner_partner campaign on February 28, 2021
romi_banner_partner = output_df.loc[output_df.index[(df['campaign_name'] == 'banner_partner') & (df['c_date'] == '2021-02-28')], 'ROMI'].iloc[0]
print("ROMI for the banner_partner campaign on February 28, 2021 is:", romi_banner_partner)
```

```python
#TASK 2 - Calculate Overall ROMI

# Calculate ROMI
total_revenue = df['revenue'].sum()
total_mark_spent = df['mark_spent'].sum()
romi = (total_revenue - total_mark_spent) / total_mark_spent

print("ROMI:", romi)
```

```python
#TASK 3 - Calculate ROMI for instagram_blogger campaign between 15th Feb to 25th Feb

# Filter data for the specified conditions
filtered_df = df[(df['campaign_name'] == 'instagram_blogger') & 
                 (df['c_date'] >= '2021-02-15') & 
                 (df['c_date'] <= '2021-02-25')]

# Calculate ROMI
romi = ((filtered_df['revenue'].sum() - filtered_df['mark_spent'].sum()) / filtered_df['mark_spent'].sum())

print("ROMI for instagram_blogger campaign between 15th Feb to 25th Feb:", romi)
```

```python
#TASK 4 - Find out the average revenue spent on weekend(Saturday and Sunday)

# Converting c_date to datetime 
df['c_date'] = pd.to_datetime(df['c_date'])

# Filtering data for Sundays and Saturdays
filtered_df = df[df['c_date'].dt.day_name().isin(['Sunday', 'Saturday'])]

# Calculate average revenue
avg_revenue = filtered_df['revenue'].mean()

print("Average revenue on Sundays and Saturdays:", avg_revenue)
```

```python
#TASK 5 - Find out the average revenue spent on weekdays(Monday to Friday) with NOT IN

# Converting c_date to datetime 
df['c_date'] = pd.to_datetime(df['c_date'])

# Filter data for days other than Sundays and Saturdays
filtered_df = df[~df['c_date'].dt.day_name().isin(['Sunday', 'Saturday'])]

# Calculate average revenue
avg_revenue = filtered_df['revenue'].mean()

print("Average revenue on days other than Sundays and Saturdays:", avg_revenue)
```

```python
#TASK 6 - Which campaign showed the worst loss in a single day? By loss we mean negative gross profit (revenue - marketing spending). Enter campaign ID.

# Selecting specific columns and calculating gross_profit
selected_df = df[['campaign_name', 'campaign_id', 'mark_spent', 'impressions', 'c_date', 'leads', 'revenue']].copy()
selected_df.loc[:, 'gross_profit'] = selected_df['revenue'] - selected_df['mark_spent']

# Finding the campaign with the worst loss in a single day
worst_loss_campaign = selected_df[selected_df['gross_profit'] == selected_df['gross_profit'].min()]

worst_loss_campaign_id = worst_loss_campaign['campaign_id'].values[0]

print("Campaign ID with the worst loss in a single day:", worst_loss_campaign_id)
```

```python
#TASK 7 - How much total money we spent on Facebook on campaigns with negative ROMI.
#Note: Use LOCATE function to find the substring in the string. 

# Filter campaigns with negative ROMI and containing 'facebook' in campaign_name
negative_romi_facebook = df[(df['ROMI'] < 0) & (df['campaign_name'].str.contains('facebook', case=False))]

# Calculate total money spent on Facebook campaigns with negative ROMI
total_spent_facebook_negative_romi = negative_romi_facebook['mark_spent'].sum()

print("Total money spent on Facebook campaigns with negative ROMI:", total_spent_facebook_negative_romi)
```
